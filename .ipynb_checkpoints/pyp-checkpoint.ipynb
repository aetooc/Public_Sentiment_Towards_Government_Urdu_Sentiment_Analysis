## Dataset Information

The objective of this task is to detect hate speech in tweets. For the sake of simplicity, we say a tweet contains hate speech if it has a racist or sexist sentiment associated with it. So, the task is to classify racist or sexist tweets from other tweets.

Formally, given a training sample of tweets and labels, where label '1' denotes the tweet is racist/sexist and label '0' denotes the tweet is not racist/sexist, your objective is to predict the labels on the test dataset.

For training the models, we provide a labelled dataset of 31,962 tweets. The dataset is provided in the form of a csv file with each line storing a tweet id, its label and the tweet.

## Preprocessing the dataset

```python
# removes pattern in the input text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
df.head()
```

```python
# remove twitter handles (@user)
df['clean_tweet'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
```

```python
df.head()
```

```python
# remove special characters, numbers and punctuations
df['clean_tweet'] = df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
df.head()
```

```python
# remove short words
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))
df.head()
```

```python
# individual words considered as tokens
tokenized_tweet = df['clean_tweet'].apply(lambda x: x.split())
tokenized_tweet.head()
```

```python
# stem the words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda sentence: [stemmer.stem(word) for word in sentence])
tokenized_tweet.head()
```

```python
# combine words into single sentence
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = " ".join(tokenized_tweet[i])
    
df['clean_tweet'] = tokenized_tweet
df.head()
```

```python
# extract the hashtag
def hashtag_extract(tweets):
    hashtags = []
    # loop words in the tweet
    for tweet in tweets:
        ht = re.findall(r"#(\w+)", tweet)
        hashtags.append(ht)
    return hashtags    
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import svm
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, Dropout, Dense, Flatten, LSTM
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import precision_score, recall_score, roc_auc_score
```

# Logical Regression

```python
# Create vectorizer for Logistic Regression
bow_vectorizer_lr = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words=stopWords)
bow_lr = bow_vectorizer_lr.fit_transform(df['clean_tweet'])
```

```python
x_train_lr, x_test_lr, y_train_lr, y_test_lr = train_test_split(bow_lr, df['label'], random_state=42, test_size=0.2)
```

```python
# Train models
model_lr = LogisticRegression()
model_lr.fit(x_train_lr, y_train_lr)
```

```python
# Make predictions and evaluate models
pred_lr = model_lr.predict(x_test_lr)
accuracy_lr = accuracy_score(y_test_lr, pred_lr)
precision_lr = precision_score(y_test_lr, pred_lr)
recall_lr = recall_score(y_test_lr, pred_lr)
f1_lr = f1_score(y_test_lr, pred_lr)
```

```python
print("Logistic Regression: \nAccuracy = {} \nPrecision = {} \nRecall = {} \nF1 Score = {}".format(accuracy_lr, precision_lr, recall_lr, f1_lr))
```

# Naive Bayes

```python
# Create vectorizer for Naive Bayes
bow_vectorizer_nb = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words=stopWords)
bow_nb = bow_vectorizer_nb.fit_transform(df['clean_tweet'])
```

```python
x_train_nb, x_test_nb, y_train_nb, y_test_nb = train_test_split(bow_nb, df['label'], random_state=42, test_size=0.2)
```

```python
model_nb = MultinomialNB()
model_nb.fit(x_train_nb, y_train_nb)
```

```python
pred_nb = model_nb.predict(x_test_nb)
accuracy_nb = accuracy_score(y_test_nb, pred_nb)
precision_nb = precision_score(y_test_nb, pred_nb)
recall_nb = recall_score(y_test_nb, pred_nb)
f1_nb = f1_score(y_test_nb, pred_nb)
```

```python
print("Naive Bayes: \nAccuracy = {} \nPrecision = {} \nRecall = {} \nF1 Score = {}".format(accuracy_nb, precision_nb, recall_nb, f1_nb))
```

```python
model_svm = svm.SVC(kernel='linear', probability=True, C=1, gamma=1)
model_svm.fit(x_train_svm, y_train_svm)
```

```python
pred_svm = model_svm.predict(x_test_svm)
accuracy_svm = accuracy_score(y_test_svm, pred_svm)
precision_svm = precision_score(y_test_svm, pred_svm)
recall_svm = recall_score(y_test_svm, pred_svm)
f1_svm = f1_score(y_test_svm, pred_svm)
```

```python
print("SVM: \nAccuracy = {} \nPrecision = {} \nRecall = {} \nF1 Score = {}".format(accuracy_svm, precision_svm, recall_svm, f1_svm))
```
